```python
##Libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

## Zadanie 1

```python
# Wczytywanie danych z pliku Excel
filename = "practice_lab_1.xlsx"
data = pd.ExcelFile(filename)
df = data.parse(sheet_name=0)  # Wczytanie pierwszego arkusza

# Pobieranie danych jako tablica numpy oraz lista nazw kolumn
values = df.to_numpy()
columns = df.columns.tolist()

# a) Różnica między tablicą z parzystymi i nieparzystymi wierszami
even_rows = values[::2, :]
odd_rows = values[1::2, :]
difference = even_rows - odd_rows

# b) Standaryzacja danych względem całej tabeli
mean_all = values.mean()
std_all = values.std()
normalized_all = (values - mean_all) / std_all

# c) Standaryzacja danych względem każdej kolumny
std_per_column = values.std(axis=0) + np.spacing(values.std(axis=0))
normalized_columns = (values - values.mean(axis=0)) / std_per_column

# d) Obliczanie współczynnika zmienności dla każdej kolumny
coefficient_of_variation = values.mean(axis=0) / std_per_column

# e) Znalezienie kolumny z największym współczynnikiem zmienności
max_cv_column = columns[np.argmax(coefficient_of_variation)]

# f) Liczba elementów większych od średniej w każdej kolumnie
count_above_mean = np.sum(values > values.mean(axis=0), axis=0)

# g) Znalezienie nazw kolumn zawierających maksymalną wartość
max_value_mask = values == values.max()
columns_with_max_value = np.array(columns)[max_value_mask.any(axis=0)].tolist()

# h) Kolumny z największą liczbą zer
zero_counts = np.sum(values == 0, axis=0)
columns_most_zeros = np.array(columns)[zero_counts == zero_counts.max()].tolist()

# i) Kolumny, w których suma wartości na pozycjach parzystych jest większa niż na nieparzystych
even_sum = values[::2, :].sum(axis=0)
odd_sum = values[1::2, :].sum(axis=0)
columns_even_greater = np.array(columns)[even_sum > odd_sum].tolist()

# Wyświetlanie wyników
print("Różnica między parzystymi i nieparzystymi wierszami:\n", difference)
print("Standaryzowane dane (dla całej tabeli):\n", normalized_all)
print("Standaryzowane dane (dla kolumn):\n", normalized_columns)
print("Współczynniki zmienności:\n", coefficient_of_variation)
print("Kolumna z największym wsp. zmienności:", max_cv_column)
print("Liczba elementów większych od średniej:\n", count_above_mean)
print("Kolumny z maksymalną wartością:", columns_with_max_value)
print("Kolumny z największą liczbą zer:", columns_most_zeros)
print("Kolumny, gdzie suma wartości na pozycjach parzystych jest większa niż na nieparzystych:", columns_even_greater)
```

## Zadanie 2

```python
## Zadanie 2: Podstawowe operacje na tablicach

# a) Utworzenie jednowymiarowej tablicy zawierającej liczby od 1 do 20
array_1d = np.arange(1, 21)

# b) Wyodrębnienie elementów od 5 do 15
subset = array_1d[4:15]

# c) Utworzenie dwuwymiarowej tablicy 5×5 z losowymi liczbami całkowitymi z przedziału 0–50
matrix_5x5 = np.random.randint(0, 51, (5, 5))

# d) Obliczenie średniej, sumy i mediany wartości tej macierzy
mean_matrix = matrix_5x5.mean()
sum_matrix = matrix_5x5.sum()
median_matrix = np.median(matrix_5x5)

# e) Mnożenie elementów dwóch tablic
matrix_1 = np.random.randint(1, 10, (3, 3))
matrix_2 = np.random.randint(1, 10, (3, 3))
matrix_product = matrix_1 * matrix_2

# Wyświetlanie wyników
print("Jednowymiarowa tablica:", array_1d)
print("Wyodrębnione elementy:", subset)
print("Macierz 5x5:\n", matrix_5x5)
print("Średnia macierzy:", mean_matrix)
print("Suma macierzy:", sum_matrix)
print("Mediana macierzy:", median_matrix)
print("Mnożenie dwóch macierzy:\n", matrix_product)
```

## Zadanie 3

```python
## Zadanie 3: Reprezentacja obrazu i modyfikacje

# a) Utworzenie obrazu 150x150 z wartościami 0
image = np.zeros((150, 150), dtype=np.uint8)

# b) Ustawienie białego kwadratu w centrum
image[50:100, 50:100] = 255

# c) Utworzenie obrazu z gradientem
gradient = np.tile(np.linspace(0, 255, 150, dtype=np.uint8), (150, 1))

# d) Obliczenie średniej wartości pikseli dla obu obrazów
mean_image = image.mean()
mean_gradient = gradient.mean()

# Wyświetlanie wyników
print("Średnia wartość pikseli w obrazie z białym kwadratem:", mean_image)
print("Średnia wartość pikseli w obrazie z gradientem:", mean_gradient)
```

```python
## Zadanie 4: Tworzenie i manipulacja DataFrame

# a) Utworzenie DataFrame
students = pd.DataFrame({
    'Student': ['Anna', 'Bartek', 'Celina', 'Damian', 'Ewa'],
    'Ocena': [78, 62, 88, 45, 90],
    'Grupa': ['A', 'B', 'A', 'B', 'A']
})

# b) Sortowanie według kolumny "Ocena" malejąco
students_sorted = students.sort_values(by='Ocena', ascending=False)

# c) Wyodrębnienie studentów z grupy "A"
students_group_A = students[students['Grupa'] == 'A']

# d) Dodanie kolumny "Status"
students['Status'] = np.where(students['Ocena'] > 70, 'Zdany', 'Niezdany')

# Wyświetlanie wyników
print("Posortowany DataFrame:\n", students_sorted)
print("Studenci z grupy A:\n", students_group_A)
print("DataFrame z kolumną 'Status':\n", students)
```

## Zadanie 5

```python
import numpy as np
import matplotlib.pyplot as plt

# Przedział wartości x
x = np.arange(-10, 10, 0.03)

# Definicje funkcji
f_tanh = np.tanh(x)
f_exp_form = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
f_sigmoid = 1 / (1 + np.exp(-x))
f_relu = np.maximum(0, x)
f_elu = np.where(x > 0, x, np.exp(x) - 1)

# Tworzenie wykresów
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.plot(x, f_tanh, label=r'$f(x) = \tanh(x)$', color='blue')
plt.legend()
plt.grid()

plt.subplot(2, 3, 2)
plt.plot(x, f_exp_form, label=r'$f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$', color='red')
plt.legend()
plt.grid()

plt.subplot(2, 3, 3)
plt.plot(x, f_sigmoid, label=r'$f(x) = \frac{1}{1+e^{-x}}$', color='green')
plt.legend()
plt.grid()

plt.subplot(2, 3, 4)
plt.plot(x, f_relu, label=r'$f(x) = \max(0, x)$', color='purple')
plt.legend()
plt.grid()

plt.subplot(2, 3, 5)
plt.plot(x, f_elu, label=r'$f(x) = x, \quad x > 0$' '\n' r'$f(x) = e^x - 1, \quad x \leq 0$', color='orange')
plt.legend()
plt.grid()

# Wyświetlenie wykresów
plt.tight_layout()
plt.show()
```
